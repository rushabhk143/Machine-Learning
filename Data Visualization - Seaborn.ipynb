```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#distplot

sns.distplot(tips['total_bill']) #The distplot shows the distribution of a univariate set of observations
```

```python
sns.distplot(tips['tip'],kde=False,bins=45) #kde=kernel density estimation
```

```python
#jointplot

sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') 
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') #kind refers to the plot shape such as scatter or hex
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') #reg will draw a regression line
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde') #default plot shape is scatter
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
#pairplot - pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).

sns.pairplot(tips) #pairplot for all the numerical values
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm') #pairplot for categorical column values #hue & #palette
```

```python
sns.rugplot(tips['total_bill'])
```

```python
#barplot - barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) #standard daviation
```

```python
#countplot - This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value:

sns.countplot(x='sex',data=tips)
```

```python
#boxplot

sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#violinplot

sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#stripplot

sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True,dodge=True) #dodge=split
```

```python
#swarmplot

sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#violinplot & swarmplot together

sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#factorplot

sns.factorplot(x='day',y='total_bill',data=tips,kind='swarm') #can change the kind to different plots forms
```

```python
#GRID

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
#map to the grid

g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal

g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#facet grid

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.hist,'total_bill') #we can change the plot types here
```

```python
#Regression Plots
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex') #lmplot=linear model plot
```

```python
# http://matplotlib.org/api/markers_api.html

#lmplot with markers

sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

```python
#using a grid

sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
#using a grid

sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=8)
```
